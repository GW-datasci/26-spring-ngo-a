```python
import pandas as pd

tweets = pd.read_csv("tweets.csv")
tweets = tweets.rename(columns={"Tweet_count": "Account"})
tweets.to_csv("tweets.csv", index=False)
print(tweets.columns.tolist())
```

```python
sample = pd.read_csv("sample_house_full.csv")

# All handles that have any row in tweets.csv (including SENTINEL rows)
scraped_handles = set(tweets["Account"].str.lower().str.strip().dropna())

# Check each sample member against that set
sample["handle_lower"] = sample["twitter"].str.lower().str.strip()
sample["scraped"] = sample["handle_lower"].isin(scraped_handles)

print(f"Total in sample : {len(sample)}")
print(f"Already scraped : {sample['scraped'].sum()}")
print(f"Not yet scraped : {(~sample['scraped']).sum()}")

# Export unscraped
not_scraped = sample[~sample["scraped"]].drop(columns=["handle_lower", "scraped"])
not_scraped.to_csv("sample_house_remaining.csv", index=False)
print("\nSaved → sample_house_remaining.csv")
print(not_scraped[["official_full", "twitter"]].to_string())
```

```python
import os
import math

os.makedirs("splitting/remaining_sample_splits", exist_ok=True)

chunk_size = math.ceil(len(not_scraped) / 4)

for i in range(4):
    chunk = not_scraped.iloc[i*chunk_size:(i+1)*chunk_size]
    path = f"splitting/remaining_sample_splits/remaining_{i+1}.csv"
    chunk.to_csv(path, index=False)
    print(f"remaining_{i+1}.csv — {len(chunk)} accounts")
```
